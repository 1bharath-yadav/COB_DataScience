# Calling public Api and creating a csv dataset 

Given Public API :https://data.binance.com/api/v3/ticker/24hr


```python
#importing requests library to communicate with webserver
#importing pandas library to frame the data
import requests
import pandas as pd
```

```python
response = requests.get("https://api.binance.com/api/v3/ticker/24hr")
Data = response.json()
```

```python
df = pd.DataFrame(Data)
```

```python
df.to_csv("binance_data" , index=False)
```
